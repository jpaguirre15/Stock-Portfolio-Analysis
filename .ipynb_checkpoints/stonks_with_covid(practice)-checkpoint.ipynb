# Risk Calculator

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from pandas_datareader import data as wb
import requests
import json
```

```python
def getStats(country):
    api_url = 'https://api.smartable.ai/coronavirus/stats/'+country
    api_params = {
        'Cache-Control': 'no-cache',
        'Subscription-Key': '3009d4ccc29e4808af1ccc25c69b4d5d',
    }
    r = requests.get(url=api_url, params=api_params) 
    return r.text

data = getStats('US')

jsonData = json.loads(data)
jsonData.keys()

country = jsonData['location']['countryOrRegion']
country

jsonData['updatedDateTime']

jsonData['stats']['breakdowns']
latestData = jsonData['stats']['breakdowns'][0]
latestData['newDeaths'] 

history = pd.DataFrame(jsonData['stats']['history'])
history['date']=pd.to_datetime(history['date'])
history=history.set_index('date')
history
```

```python
history.info()
```

```python
history.plot(figsize=(10,5), title=country)
```

```python
# Tickers
tickers = ['AMZN', 'BABA', 'AZN', 'INTC', 'AAPL', 'TCEHY', 'PFE', 'DAL', 'CCL']
my_portfolio = pd.DataFrame()
for t in tickers:
    my_portfolio[t] = wb.DataReader(t, data_source='yahoo', start='2020-01-22')['Adj Close']
```

```python
my_portfolio.info()
```

```python
my_portfolio.tail()
```

```python
(my_portfolio/my_portfolio.iloc[0]*100).plot(figsize=(15,6))
```

```python
# Concatinate COVID-19 cases with Current Portfolio
combin = pd.concat([history, my_portfolio], axis=1, join='inner')
combin = pd.concat([history, my_portfolio.reindex(history.index)], axis=1)
combin
```

```python
combin.info()
```

```python
# Covariance 
combin_cov = combin.cov()
combin_cov
```

```python
# Correlations between COVID-19 and portfolio
combin_corr = combin.corr()
combin_corr
```

```python
sec_returns = np.log(my_portfolio/my_portfolio.shift(1))
sec_returns
```

```python
# Concatinate COVID-19 cases with Current Portfolio Returns
combin_return = pd.concat([history, sec_returns], axis=1, join='inner')
combin_return = pd.concat([history, sec_returns.reindex(history.index)], axis=1)
combin_return
```

```python
# Covariance 
combin_return_cov = combin_return.cov()
combin_return_cov
```

```python
# Correlations between COVID-19 and portfolio
combin_return_corr = combin_return.corr()
combin_return_corr
```
